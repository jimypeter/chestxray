```python
from fastai.vision import *
```

```python
learn = load_learner('.')
```

```python
model = learn.model.eval()
```

```python
image_sz = 256
```

```python
dummy_data = torch.randn(1, 3, image_sz, image_sz)
```

```python
dummy_input = torch.autograd.Variable(dummy_data)
```

```python
model_name = 'pneumonia.onnx'
input_name = 'image'
output_name = 'diagnosis'
```

```python
torch.onnx.export(model, dummy_input,
                  model_name, input_names = [input_name], output_names = [output_name])
```

```python
# optionally, check if the model is ok
import onnx
onnx_model = onnx.load(model_name)
onnx.checker.check_model(onnx_model)
```
