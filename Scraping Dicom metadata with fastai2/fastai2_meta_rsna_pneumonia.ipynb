```python
from fastai2.basics import *
from fastai2.medical.imaging import *
```

```python
path = Path('.')

path_trn = path / 'stage_2_train_images'
path_tst = path / 'stage_2_test_images'
fns_trn = path_trn.ls()
fns_tst = path_tst.ls()
len(fns_trn),len(fns_tst)
```

```python
df_tst = pd.DataFrame.from_dicoms(fns_tst, n_workers=8, px_summ=True)
```

```python
df_trn = pd.DataFrame.from_dicoms(fns_trn, n_workers=8, px_summ=True)
```

```python
df_tst.to_csv('df_tst.csv')
df_trn.to_csv('df_trn.csv')
```
